## 3. Univariate Analysis

```python
# a. Categorical Features

# Define columns for pie and bar chart
pie_chart_cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService','PaperlessBilling']
bar_chart_cols = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                  'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaymentMethod']
# Combine all for layout planning
all_cols = pie_chart_cols + bar_chart_cols
n_cols = 4
n_rows = (len(all_cols) + n_cols - 1) // n_cols
#Adjust figure size
fig, axes = plt.subplots(n_rows, n_cols, figsize = (24, n_rows * 5))
axes = axes.flatten()
# Prepare pie chart data
for i, col in enumerate(all_cols):
    ax = axes[i]
    if col in pie_chart_cols:
        #Prepare pie chart data
        values = df[col].value_counts()
        ax.pie(values, labels=values.index, 
autopct='%1.1f%%', startangle=90, 
colors=sns.color_palette('pastel'))
        ax.set_title(f'{col} Distribution (Pie)', fontsize=11)
    else:
# COunt plot with churn hue  
        sns.countplot(data=df, x=col, hue='Churn', ax=ax)
        ax.set_title(f'Churn by {col}')
        ax.tick_params(axis='x', rotation=45)
# Remove any empty subplots        
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
# Adding spacing between rows
plt.tight_layout
plt.subplots_adjust(hspace=0.7, wspace=0.2)

plt.show()
```

```python
# b. Numerical Features
numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
for col in numerical_cols:
    plt.figure(figsize=(6, 5))
    sns.histplot(data=df, x=col, hue='Churn', kde=True)
    plt.title(f'Distribution of {col} by Churn')
    plt.tight_layout()
    plt.show()
```

## 4. Bivariate Analysis

```python
# a. Churn by Contract Type
contract_churn = df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()
contract_churn.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Proportion')
plt.show()
```

```python
# b. Tenure  vs MonthlyCharges
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='tenure', y='MonthlyCharges', hue='Churn')
plt.title('Tenure vs Monthly Charges by Churn')
plt.show()
```

## 5. Correlation Matrix

```python
# Encode churn for correlation
df_encoded = df.copy()
df_encoded['Churn'] = df_encoded['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
# Select numerical features
corr_matrix = df_encoded[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']].corr()
```

```python
# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Key Insights
    1. Shorter tenure and month-to-month contracts have higher churn.
    2. Paperless billing customers are more likely to churn.
    3. Customers without tech support or online security are at risk.
    4. High monthly charges with little tenure can signal churn
