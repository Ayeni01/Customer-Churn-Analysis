## TASK: Visualize Key Insights with Seaborn
### Objective:
Use Seaborn and Matplotlib to visualize churn-related trends, model performance, and important feature relationships.


### 1. Setup

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Load the dataset again if needed
df = pd.read_csv('telco_cleaned_for_analysis.csv')
# Set visual style
sns.set(style='whitegrid')
```

### 2. Churn Distribution

```python
sns.countplot(data=df, x='Churn', palette='coolwarm')
plt.title('Customer Churn Distribution')
plt.show()
```

### 3. Tenure and Churn

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='tenure', hue='Churn', kde=True, multiple='stack')
plt.title('Tenure Distribution by Churn')
plt.xlabel('Tenure (Months)')
plt.ylabel('Number of Customers')
plt.show()
```

### 4. Monthly Charges vs. Churn

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette='Set2')
plt.title('Monthly Charges vs. Churn')
plt.show()
```

### 5. Contract Type and Churn

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Contract', hue='Churn', palette='magma')
plt.title('Churn Rate by Contract Type')
plt.xticks(rotation=30)
plt.show()
```

### 6. Internet Service Type vs. Churn


```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='InternetService', hue='Churn', palette='Set1')
plt.title('Churn by Internet Service Type')
plt.xticks(rotation=30)
plt.show()
```

### 7. Churn Heatmap of Key Features
     To explore churn correlations with selected encoded features:

```python
df_corr = df.copy()
df_corr['SeniorCitizen'] = df_corr['SeniorCitizen'].apply(lambda x: 1 if x == 'Yes' else 0)
df_corr['Churn'] = df_corr['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)
corr = df_corr[['Churn', 'tenure', 'MonthlyCharges', 'TotalCharges', 'SeniorCitizen']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Key Takeaways from Visualizations
    Tenure and contract type have strong visual patterns with churn.
    Month-to-month contracts clearly have higher churn.
    Higher monthly charges combined with short tenure are red flags.
    Heatmaps reveal the correlation strength for numeric churn predictors.
